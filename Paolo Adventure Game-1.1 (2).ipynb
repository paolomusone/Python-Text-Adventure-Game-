```python
"""
Created on Wednasday January 26 2022

@author: Paolo Musone

"""

"""
    DocString:
    
    A) Introduction:
    This game is based on the famous Tv show "Breaking Bad"! 
    In this adventure game the player will experience being
    Walter White, the main character, with the difference that 
    he will have the choice to rewrite part of the
    original story. 
    
    The game consist of three levels:
    
    Level 1: Decide if you want to start cooking methaphetamine
    Level 2: How to kill Tuco
    Level 3: Questions by Gus Fring
    
    
    In the following link you will find some useful information about this TV Show.
    https://en.wikipedia.org/wiki/Breaking_Bad#:~:text=12%20External%20links-,Premise,diagnosed%20with%20terminal%20lung%20cancer.
    
    
    B) Known Bugs and/or Errors:
    None.  
    
"""
#importing packages
import time
import random
import sys


#defining the first function to start the game
def welcome():
    
    #printing the introduction text
    print( """    Hey you! Welcome to the Breaking Bad world! 
    Today in this game you will go through the story of Breaking bad!
    But different from the actual Tv Show. You will be able to change the story 
    because you are going to be the main character, Walter White!
    Ok! Let's begin!""")
    
    #stop the flow for 0.5 seconds
    time.sleep(0.5)
    
    #press enter to continue the game
    input(prompt="<Press enter to continue>\n")
    
    #stop the flow for 0.5 seconds
    time.sleep(0.5)
    
    #asking the player's name through the "name" variable
    name=input(prompt="<Please enter your name>\n")
    name= name.title()#Capitalizing the initial letters of his/her name
    
    #printing the second introduction with the player's name.
    print(f"""
    Hi, {name} White! You are a chemistry teacher in Albequerque,
    a very respectable person but a little bit bored of your routine. 
    Unfortunately you just got to know that you are very sick and there is not 
    a lot of time remaining for you. Doctors say about 1 year.

    Since you want to take care of your family, financially, 
    you are actually starting to think to make some money
    by cooking methaphetamine with your ex student Jesse Pinkman! 
    What a good idea!""")
    first()#move the player to the "first" function
     
##########################################################
#defining the function
def first():
    #asking the first question to the player
    question1 = input(prompt=f"""
         Do you want to start 'cooking'? 
         1) YES!
         2) NO!
        \n""").lower() 
    
  
    #if/else/elif conditional statement for question1
    if "1" in question1  or "y" in question1:
        print("\n That's the spirit! You are going to make your family rich!\n")
        input(prompt="<Press enter to continue>\n")
        second()#moving the player to the "second" function
    
    elif "2" in question1 or "n" in question1:
        print("""Boring! Come on, lets give it a try! Try again!\n""")
        input(prompt="<Press enter to continue>\n")
        first() #moving the player to the "first" function
    
    #typo. Give the player another chance to answer correctly
    else:
        print(""" Unexpected error!Feel free to run the code again and write only one time
        one of the following number/words:
        1
        2
        yes
        no.\n""")
        input(prompt="<Press enter to continue>\n")
        first() #moving the player to the "first" function

################################################################
#definining the function
def second():
    
    #printing another piece of the story
    print("""
        Now you are finally cooking with Jesse Pinkamn and selling your product to a drug dealer
        called Tuco. he is a very dangerous and irrational person; 
        really not the type of guy you want to go into business with!
        
        Jesse, your coworker, is actually thinking to kill him and he is asking you to make a plan.""")
    
    #asking the second question to the player
    question2 = input(prompt="""
        How do you want to kill Tuco?
        1) using ricin; a lethel poison
        2) shooting him with a gun
        3) running a car over him
        4) you don't want to kill Tuco\n """).lower()
    
    #if/else/elif conditional statement for question2
    if "1" in question2  or "ricin" in question2 or "poison" in question2 :
        
        print(""" 
        It worked! Ricin has done the job perfectly!Tuco is not a problem anymore!  
        But now you have to find a new drug lord to sell your product to!
        Remember, you want to make money!""")
        input(prompt="<Press enter to continue>\n")
        third()#moving the player to the "third" function
    
        
    elif "2" in question2 or "shoot" in question2 or "gun" in question2 :
        print("""
        It worked!You just made a perfect headshot to Tuco! He is not a problem anymore! 
        But now you have to find a new drug lord to sell your product to!
        Remember, you want to make money!""")
        input(prompt="<Press enter to continue>\n")
        third()#moving the player to the "third" function
        
    #give the player another chance to continue the game without interruption   
    elif "3" in question2 or "car" in question2 or "running" in question2 or "over" in question2 :
        print("""Running over him with a car it dindt actually work!
              Tuco saw you coming with the car with the intention to kill him 
              and he managed to run away. Now is looking for revenge! 
              But you have another chance!
              Please try another way to kill Tuco. """)
        input(prompt="<Press enter to continue>\n")
        second()#moving the player to the "second" function
        
    #wrong answer.The player has lost the game    
    elif "4" in question2 or "kill" in question2 or "don't" in question2 or "you" in question2 :
        print(f"""Tuco eventually killed you. You knew that he was a dangerous man!""")
        fail()#moving the player to the "fail" function
        
    #typo. Give the player another chance to answer correctly. 
    else :
        print("""
        I am sorry, I did not understand! Please try something different
        and remember to pick one of the available options""")
        input(prompt="<Press enter to continue>\n")
        second()
#########################################################
#defining the function
def third():
    print("""
    There is a potential drug lord , Gus Fring, very interested in your product
    but is very diffident and he wants to test you before setting a deal!
    You arranged to meet him and now he is about to ask you a question.""")
    #asking the third question to the player
    
    'Do you trust your partner: Jessy Pinkman?'
    #question list
    question_list = ['Are you a cop?',
                     'Do you trust your partner: Jessy Pinkman?',
                     'Do you have a family?']
    
    questions = 1 # variable for the while loop
    
    
    #While loop
    while questions > 0:
        
        #asking the third question to the player
        question = random.choice(question_list)
        
        input(prompt="<Press enter for your question>\n")
        print(question)
        print(" ")
        
        #if/else/elif conditional statement for the loop
        if question == question_list[0]: 
            answer1 = input(prompt="Answer\n").lower()
            if "n" in answer1 or "not" in answer1 or "never" in answer1 :
                print("I am glad to hear that! You are hired!")
                questions -=1#break the while loop
                win()#moving the player to the "win" function
            
            else:
                print("After this answer how am i supposed to trust you?")
                questions -=1#break the while loop
                fail()#moving the player to the "fail" function
            
            
            
        elif question == question_list[1]:
            answer2 = input(prompt="Answer\n").lower()
            if "yes" in answer2 or "y" in answer2 or "do" in answer2 or "absoluty"in answer2:
                print("""Good! I was actually excpecting this answer! Let's start with the business""")
                questions -=1 #break the while loop
                win()#moving the player to the "win" function
                    
            else:
                print("""what kind of person are you If you do not even trust your partner Jessy
                       There is no reason for me to continue this meeting! """)
                fail()#moving the player to the "fail" function
                questions -=1#break the while loop
                
        elif question == question_list[2]:
            answer3 = input(prompt="Answer\n").lower()
            if "yes" in answer3 or "y" in answer3 or "do" in answer3 or "absoluty" in answer3:
                print("""Good! That makes you a responsible person! Now I want to do business with you  """)
                questions -=1#break the while loop
                win()#moving the player to the "win" function
                
                
            else:
                print("""So you are without a family? 
                  That makes it easier for you to betray me! We do not have a deal!""")
                fail()#moving the player to the "fail" function
                
        #typo. Give the player another chance to answer correctly       
        else:
            print("I did not expect this answer! Could you please try again?")
            questions -=1#break the while loop
            third()#moving the player to the "third" function
            
#define the fail function
def fail():
    
    #printing the end of the game in the case the player fails 
    print("\n Game Over!.")
    print("\n I am sorry but you lost the game.")
    time.sleep(0.5)#stop the flow for 0.5 seconds
    restart=input(prompt='<Do you wish to play again?>').lower()
    
    if "y" in restart or "k" in restart or "sure" in restart:
        print("Good idea!")
        time.sleep(0.5) #stop the flow for 0.5 seconds
        welcome()#moving the player to the "welcome" function, "start of the game"
        
    elif "n" in restart:
        
        #to end the code
        try: 
            print("\n Good Bye! \n")
            import sys
            sys.exit()#quit the game
            
        #alternative solution for try 
        except: 
            quit()#quit the game
            
    #typo. Give the player another chance to answer correctly
    else:
        print("\n Unexpected error. Please make sure to answer the question with yes or no \n")
        time.sleep(0.5)#stop the flow for 0.5 seconds
        fail() #moving the player to the "fail" function
    
#define the win function
def win():
    
    #printing the end of the game in the case the player wins.
    print("""Congratulation! You won the Game!You are one step to become one of the most notorious drug lords, the 'Heisenberg'!""")
    time.sleep(0.5)#stop the flow for 0.5 seconds
    restart1=input(prompt='<Do you wish to play again?>').lower()
    
    if "y" in restart1 or "k" in restart1 or "sure" in restart1 :
        print("Good idea!")
        time.sleep(0.5)#stop the flow for 0.5 seconds
        welcome()#moving the player to the "welcome" function, "start of the game"
        
    elif "n" in restart1:
        
        #to end the code
        try: 
            print("\n Good Bye \n")
            import sys
            sys.exit()#quit the game
            
        #alternative solution for try    
        except: 
            quit()#quit the game
            
    #typo. Give the player another chance to answer correctly
    else:
        print("\n Unexpected error. Please make sure to answer the question with yes or no \n")
        time.sleep(0.5)#stop the flow for 0.5 seconds
        win()#moving the player to the "win" function 

#starting the game 
welcome() 



    

  
    
    
   

```
